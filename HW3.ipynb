```python
url="https://raw.githubusercontent.com/gjm112/DSCI401/main/data/Pitching.csv"
url2="https://raw.githubusercontent.com/gjm112/DSCI401/ac29f11ebae811bee9be4bf71505e514e951a4e9/data/Batting.csv"
url3="https://raw.githubusercontent.com/gjm112/DSCI401/main/data/People.csv"
```

```python
data_main=pd.read_csv(url3)
data_bat=pd.read_csv(url2)
data_pit=pd.read_csv(url)
```

#Solution of 1(a):
 groupby playerID and take sum of SB and HR and filter values>=300 and inner join with people 

```python
data_bat1=data_bat[["playerID","HR","SB"]].groupby("playerID")[["HR","SB"]].sum().dropna()
df_batting=data_bat1.loc[(data_bat1["HR"]>=300) & (data_bat1["SB"]>=300)].reset_index()
df_1a=pd.merge(data_main, df_batting, on="playerID", how="inner")
df_1a["name_list"]=(df_1a["nameFirst"])+" "+(df_1a["nameLast"])
df_1a["name_list"].reset_index(drop=True)
```

#solution of 1(b)
similar to 1(a) but using pitching data

```python
data_pit1=data_pit[["playerID","W","SO"]].groupby("playerID")[["W","SO"]].sum().dropna()
data_pitching=data_pit1.loc[(data_pit1["W"]>=300)&(data_pit1["SO"]>=3000)].reset_index()
df_1b=pd.merge(data_main, data_pitching, on="playerID", how= "inner")
df_1b["name_list"]=(df_1b["nameFirst"])+" "+ (df_1b["nameLast"])
df_1b["name_list"].reset_index(drop=True)
```

#solution of 1(c):


```python
data_batc=data_bat[["playerID","yearID","AB","H","HR"]].groupby(["playerID","yearID"]).sum(["HR", "AB","H"]).reset_index()
df_battingc=data_batc.loc[data_batc["HR"]>=50].reset_index()
df_1c=pd.merge(data_main, df_battingc, on="playerID", how="inner")
df_1c["name_list"]=df_1c["nameFirst"]+' '+df_1c["nameLast"]
df_1cc=df_1c[["name_list","yearID","HR", "AB", "H"]].sort_values(["name_list","yearID"]).reset_index(drop=True)
df_1cc["batting_avg"]=(df_1cc["H"])/df_1cc["AB"].round(2)
df_1cc.groupby(["yearID","name_list"]).min("batting_avge").round(2).reset_index()
df_1cc.rename(columns={"name_list": "full_Name", "HR": "total_HR", "AB": "total_at_bat","H":"total_hit","batting_avg":"total_hit/total_at_bat" }).round(2)
```
